## 3. What should we do with the customer_id column?

```python
customer_df=customer_df.drop(['Customer','Effective To Date'], axis=1)
```

## 4. Load the continuous and discrete variables into numericals_df and categorical_df variables, for eg.:
- numerical_df = customer_df.select_dtypes()
- categorical_df = customer_df.select_dtypes()

```python
numerical_df = customer_df.select_dtypes(['float64', 'int64'])
categorical_df = customer_df.select_dtypes('object')
```

```python
categorical_df.head()
```

## 5. Plot every categorical variable. What can you see in the plots? Note that in the previous lab you used a bar plot to plot categorical data, with each unique category in the column on the x-axis and an appropriate measure on the y-axis. However, this time you will try a different plot.
```python
categorical_df.head()
```

```python
numerical_df.head()
```

```python
for col in categorical_df.columns:
    plt.figure(figsize=(10, 4))
    sns.barplot(x=col, y='Number of Open Complaints', data=customer_df, estimator=np.sum)
    plt.title(f'Barplot of Total Number of Open Complaints per {col}')
    plt.xticks(rotation=90)  # This will rotate the labels on the x-axis to make them more readable
    plt.show()
```

```python
for col in categorical_df.columns:
    plt.figure(figsize=(10, 4))
    sns.barplot(x=col, y='Number of Open Complaints', data=customer_df, estimator=np.average)
    plt.title(f'Barplot of Average Number of Open Complaints per {col}')
    plt.xticks(rotation=90)  # This will rotate the labels on the x-axis to make them more readable
    plt.show()
```

## 6. For the categorical data, check if there is any data cleaning that need to perform. Hint: You can use the function value_counts() on each of the categorical columns and check the representation of different categories in each column. Discuss if this information might in some way be used for data cleaning.

```python
for col in categorical_df.columns:
    print(f"Column Name: {col}")
    print(f"Unique Values Count: {categorical_df[col].nunique()}")
    print(f"Value Counts: ")
    print(categorical_df[col].value_counts())
    print("\n" + "-"*50 + "\n")
```

```python
# only Policy column need to be cleaned
customer_df[['Policy_Type','Policy_Level']] = customer_df['Policy'].str.split(' ',expand=True)
one_hot = pd.get_dummies(customer_df['Policy'])
customer_df = customer_df.drop('Policy',axis = 1)
customer_df = customer_df.join(one_hot)
```
